**1. O arquivo TraceMeast.mat contém uma lista de medidas de metais traço hipotéticos extraídos de um testemunho de sedimento. Acredita-se que a amostra de sedimentos contém um determinado numero de minerais diferentes. Use a ACP para calcular quantos minerais existem nesta amostra (as componentes principais), a abundancia relativa de cada um deles (os respectivos autovalores) e acomposição de cada componente (a carga fatorial).**

```python
# Importando as bibliotecas
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from scipy.io import loadmat
```

```python
# Importando e verificando o tipo de dado
data = loadmat('TraceMeast.mat')
print(type(data))
```

```python
# Identificando o tipo de cada chave no dicionario
for key in data.keys():
    print(key, "\t", type(data[key]))
```

```python
# Acessando os dados (chave 'X')
print(data.get('X'))
```

```python
# Associando os dados a variável data
data = data.get('X')
```

```python
# Dimensões do dado
data.shape
```

```python
# PCA
pca = PCA(n_components = 6)
pca.fit(data)
```

```python
# Autovalores
autoval = np.round(pca.explained_variance_, 1)
print('Autovalores')
print(autoval)
```

```python
# Autovetores
autovet = pca.components_
print('Autovetores')
print(autovet)
```

```python
# Variância explicada
var_exp = pca.explained_variance_ratio_ * 100
var_exp = np.round(var_exp, 2)
print('Variância explicada %:')
print(var_exp)
```

```python
# Variância explicada acumulada
var_exp_cum = np.cumsum(var_exp)
print('Variância explicada acumulada')
print(var_exp_cum)
```

```python
# Carga Fatorial
AR = autovet * (autoval * 0.5)
print('Carga Fatorial:')
print(np.round(AR, 2))
```

```python
# Tabela com os resultados
cp = ['PC %s' %i for i in range(1, len(autoval)+1)]
df = pd.DataFrame({'autovalores': autoval,
                      'var_exp_cum': var_exp_cum,
                      'var_exp': var_exp,
                      'Componente': cp})
df
```

**2. O arquivo HypoGyre.mat consiste em uma serie temporal de dados artificiais de altura do nível do mar organizados na forma de mapas tridimensionais (Lon, Lat, tempo). Converta os dados na forma adequada para o calculo das EOFs, plote as principais EOFs assim como a sua variação temporal. Interprete o resultado obtido da forma que julgar conveniente.**

```python
# Carregando os dados
data = loadmat('HypoGyre.mat')
```

```python
# Identificando o tipo de cada chave no dicionario
for key in data.keys():
    print(key, '\t', data[key])
```

```python
# Acessando os dados
lon = data.get('x') 
lat = data.get('y') 
time = data.get('t') 
ssh = data.get('Z') 
```

```python
# Dimensões dos dados
print('Latitude')
print(lat.shape)
print('-='*15)
print('Longitude')
print(lon.shape)
print('-='*15)
print('Tempo')
print(time.shape)
print('-='*15)
print('SSH')
print(ssh.shape)
```

```python
# Criando a matriz N (tempo) x M (espaço) [24x900]
dados = ssh.ravel() # [21000,1]
dados = dados.reshape(900,24) # [900,24]
dados = dados.T
dados.shape
```

```python
# Criando DataFrame dos dados
z = pd.DataFrame(dados)
z.head()
```

```python
# Matriz covariância
dados_t = dados.T
ssh_cov = np.cov(dados_t)
print('Matriz covariância:')
print(ssh_cov)
```

```python
# Autovalores e  autovetores
auto_val, auto_vec = np.linalg.eig(ssh_cov)
idx = auto_val.argsort()[::-1]
auto_val = auto_val[idx]
auto_vec = auto_vec[:,idx]
```

```python
# Autovalores
print('Autovalores')
print(np.round(auto_val,2))
```

```python
# Autovetores
print('Autovetores')
print(np.round(auto_vec,2))
```

```python
# Carga Fatorial
ar = auto_vec * (auto_val * 0.5)
print('Carga fatorial')
print(np.round(ar,2))
```
